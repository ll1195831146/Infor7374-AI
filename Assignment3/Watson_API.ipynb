```python
with open('Complete.json') as train_f:
   data = json.load(train_f)
```

```python
import numpy as np
import pandas as pd
data = pd.DataFrame(data)
```

```python
data = data.drop(['sentiment'], axis=1)
```

```python
idx = 0
for i in data['text']:
    if len(i) < 50:
        data.drop(data.index[idx], inplace = True)
    else:
        idx = idx + 1
        
```

```python
score = 0 
# a = 0
for i in data['text']:
    response = natural_language_understanding.analyze(
    text= i ,
    features=Features(
        entities=EntitiesOptions(emotion=False, sentiment=True, limit=2),
        keywords=KeywordsOptions(emotion=False, sentiment=True,
                                 limit=2))).get_result()
#     print(a)
#     a += 1 

    print(json.dumps(response, indent=2))
    if response['entities']:
        score = score + response['entities'][0]['sentiment']['score']
    elif response['keywords']:
        score = score + response['keywords'][0]['sentiment']['score']
    else:
        score += 0
    
```
